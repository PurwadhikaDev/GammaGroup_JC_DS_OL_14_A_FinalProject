Konsekuensi : Membuang biaya retensi yang tidak diperlukan seperti memberikan diskon, penawaran khusus, dll kepada pelanggan yang tidak berencana untuk churn.

### Feature Selection

Chi-Square for Categorical Variables

```python
from scipy.stats import chi2_contingency

# Define function for chi-square test
def chi_square_test(df, cols):
    results = {}
    for col in cols:
        contingency_table = pd.crosstab(df[col], df['Churn'])
        chi2, p, dof, expected = chi2_contingency(contingency_table)
        results[col] = p
    return results

# Perform chi-square test for each categorical variable
categorical_vars = ['Partner', 'Dependents', 'PhoneService', 'InternetService',
                       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
                       'StreamingMovies', 'Contract', 'PaymentMethod']
results_chi2 = chi_square_test(df, categorical_vars)

# Print results
print("Chi-Square Test Results:")
for var, p_value in results_chi2.items():
    if p_value < 0.05:
        print(f"There is a significant association between {var} and Churn (p-value: {p_value})")
    else:
        print(f"No significant association between {var} and Churn (p-value: {p_value})")
```

T-test for Numerical Variables

```python
from scipy.stats import ttest_ind

# Define function for t-test
def t_test(df, col):
    churned = df[df['Churn'] == 'Yes'][col]
    not_churned = df[df['Churn'] == 'No'][col]
    t_stat, p = ttest_ind(churned, not_churned, equal_var=False)  # Welch's t-test
    return p

# List of numerical columns for the t-test
numerical_cols = ['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges']

# Perform t-test for each numerical variable
results_ttest = {}
for var in numerical_cols:
    p_value = t_test(df, var)
    results_ttest[var] = p_value

# Print results
print("\nT-Test Results:")
for var, p_value in results_ttest.items():
    if p_value < 0.05:
        print(f"There is a significant difference in {var} between churned and non-churned customers (p-value: {p_value})")
    else:
        print(f"No significant difference in {var} between churned and non-churned customers (p-value: {p_value})")
```

### Model Benchmarking : K-Fold

K-fold cross-validation merupakan metode yang digunakan untuk mengevaluasi model dengan membagi dataset menjadi K bagian yang sama besar. Model dilatih K kali, di mana setiap kali satu bagian digunakan sebagai data test, dan sisanya sebagai data train. Hasil dari setiap iterasi dihitung rata-ratanya untuk memberikan estimasi performa model yang lebih stabil dan mengurangi bias serta menghindari overfitting.

```python
# list model yang akan di-benchmark
models = [logreg,knn,dt,rf,xgb,lgbm]

# List untuk menyimpan nilai
score=[]
rata=[]
std=[]

# Iterasi untuk setiap model
for i in models:
    skfold=StratifiedKFold(n_splits=5)      # Inisialisasi StratifiedKFold dengan 5 fold
    estimator=Pipeline([
        ('preprocess',transformer),         # Menyisipkan tahap preprocessing (OneHotEncoder) dalam pipeline
        ('model',i)])                       # Menambahkan model klasifikasi ke dalam pipeline

    # Melakukan cross-validation dan menghitung skor menggunakan Recall
    model_cv=cross_val_score(estimator,X_train,y_train,cv=skfold,scoring='recall')

    # Menyimpan nilai
    score.append(model_cv)
    rata.append(model_cv.mean())
    std.append(model_cv.std())

# Membuat DataFrame untuk menampilkan model dan skor
pd.DataFrame({'model':['Logistic Regression', 'KNN', 'Decision Tree', 'Random Forest', 'XGBoost', 'LightGBM'],'mean recall':rata,'sdev':std}).set_index('model').sort_values(by='mean recall',ascending=False)
```


Hasil benchmarking menggunakan k-fold cross-validation pada default parameter menunjukkan bahwa nilai mean recall tertinggi didapatkan oleh model LightGBM dengan nilai 0.56 dan diikuti oleh Logistic Regression dengan mean recall 0.55.

Model benchmarking ini menggunakan data yang sudah dibagi menjadi training dan test sets secara terpisah. Setiap model dilatih pada data training dan kemudian akan dievaluasi pada data test dengan menghitung Recall score.

### Test Oversampling with K-Fold Cross Validation

Untuk meningkatkan hasil model Logistic Regression, diterapkan teknik oversampling menggunakan K-Fold Cross Validation dengan membagi data menjadi K fold. Oversampling dilakukan pada data training di setiap iterasi. Model dilatih pada training set yang telah di-oversample dan diuji pada data test.

Oversampling bertujuan untuk memperbaiki kinerja model terutama dalam menangani imbalance data, sehingga model yang dihasilkan tidak bias dan dapat memprediksi kedua kelas dengan lebih akurat.

Teknik oversampling ini menambahkan data dari kelas minoritas menggunakan `RandomOverSampler`, yang berfungsi untuk menyeimbangkan distribusi kelas dalam data training. Dengan demikian, model memiliki kesempatan yang lebih baik untuk mempelajari pola pada kelas minoritas dan meningkatkan kemampuannya dalam memprediksi kedua kelas secara adil.

```python
# Menghitung Metrik Evaluasi untuk Model pada Data Training
def calc_train_error(X_train, y_train, model):
    '''returns in-sample error for already fit model.'''
    predictions = model.predict(X_train)                # Menggunakan model untuk memprediksi label dari data fitur training
    predictProba = model.predict_proba(X_train)         # Menghitung probabilitas prediksi dari data pelatihan

    accuracy = accuracy_score(y_train, predictions)             # Menghitung akurasi model (membandingkan prediksi dengan label asli)
    f1 = f1_score(y_train, predictions, average='macro')        # Menghitung skor F1

    roc_auc = roc_auc_score(y_train, predictProba[:,1], multi_class='ovr')         # Menghitung AUC-ROC:mengukur kemampuan model dalam membedakan kelas
    recall = recall_score(y_train, predictions, pos_label=1)                       # Menghitung Recall
    precision = precision_score(y_train, predictions, pos_label=1)                 # Menghitung Precision
    report = classification_report(y_train, predictions)                           # Menhasilkan report dari klasifikasi

    # Mengembalikan hasil evaluasi
    return {
        'report': report,
        'f1' : f1,
        'roc': roc_auc,
        'accuracy': accuracy,
        'recall': recall,
        'precision': precision
    }

# Menghitung Metrik Evaluasi untuk Model pada Data Testing
def calc_validation_error(X_test, y_test, model):
    '''returns out-of-sample error for already fit model.'''
    predictions = model.predict(X_test)
    predictProba = model.predict_proba(X_test)
    accuracy = accuracy_score(y_test, predictions)
    f1 = f1_score(y_test, predictions, average='macro')
    roc_auc = roc_auc_score(y_test, predictProba[:,1], multi_class='ovr')
    recall = recall_score(y_test, predictions, pos_label=1)
    precision = precision_score(y_test, predictions, pos_label=1)
    report = classification_report(y_test, predictions)
    return {
        'report': report,
        'f1' : f1,
        'roc': roc_auc,
        'accuracy': accuracy,
        'recall': recall,
        'precision': precision
    }

# Melatih Model dan Menghitung Metrik Kesalahan untuk Data Training dan Testing
def calc_metrics(X_train, y_train, X_test, y_test, model):
#     '''fits model and returns the in-sample error and out-of-sample error'''
    model.fit(X_train, y_train)         # fitting model
    train_error = calc_train_error(X_train, y_train, model)         # Menghitung error pada data training
    validation_error = calc_validation_error(X_test, y_test, model)     # Menghitung error pada data testing
    return train_error, validation_error            # Mengembalikan nilai
```

```python
from sklearn.model_selection import StratifiedKFold

K = 10          # Jumlah fold untuk K-Fold CV
kf = StratifiedKFold(n_splits=K, shuffle=True, random_state=16)         # Inisialisasi StratifiedKFold
```

```python
# Mendefinisikan variabel
data = X_train
target = y_train
```

```python
# List untuk menyimpan error tanpa oversampling
train_errors_without_oversampling = []
validation_errors_without_oversampling = []

# List untuk menyimpan error dengan oversampling
train_errors_with_oversampling = []
validation_errors_with_oversampling = []

# Melakukan iterasi untuk setiap fold
for train_index, val_index in kf.split(data, target):

    # Memisahkan data menjadi data train dan validation
    X_train, X_val = data.iloc[train_index], data.iloc[val_index]
    y_train, Y_val = target.iloc[train_index], target.iloc[val_index]

    # Inisialisasi RandomOverSampler untuk oversampling
    ros = RandomOverSampler()

    # Melakukan oversapmling pada data train
    X_ros, Y_ros = ros.fit_resample(X_train, y_train)

    # Inisialisasi model Logistic Regression
    logreg = LogisticRegression(solver='liblinear', random_state=16)         # Menggunakan solver 'liblinear' untuk klasifikasi biner
    estimator = Pipeline([                  # Pipeline dengan preprocessing dan model
        ('preprocess', transformer),
        ('model', logreg)
    ])

    # Menghitung error
    train_error_without_oversampling, val_error_without_oversampling = calc_metrics(X_train, y_train, X_val, Y_val, estimator)
    train_error_with_oversampling, val_error_with_oversampling = calc_metrics(X_ros, Y_ros, X_val, Y_val, estimator)

    # Menambahkan error kedalam list
    train_errors_without_oversampling.append(train_error_without_oversampling)
    validation_errors_without_oversampling.append(val_error_without_oversampling)

    train_errors_with_oversampling.append(train_error_with_oversampling)
    validation_errors_with_oversampling.append(val_error_with_oversampling)
```

### Hyperparameter Tuning

Hyperparameter tuning merupakan proses menemukan kombinasi terbaik dari hyperparameter yang digunakan oleh model machine learning. Hyperparameter adalah parameter yang ditentukan sebelum proses pelatihan dimulai.

Proses tuning dilakukan dengan mengeksplorasi berbagai kombinasi hyperparameter melalui teknik GridSearchCV, yang akan melakukan pencarian di seluruh ruang hyperparameter yang telah didefinisikan, dan mengevaluasi performa model dengan menggunakan teknik cross-validation. Dengan cara ini akan didapatkan kombinasi hyperparameter yang memberikan hasil terbaik.

Hyperparameter tunung bertujuan untuk meningkatkan performa model dengan memastikan bahwa model tidak overfitting atau underfitting terhadap data.

```python
# Inisialisasi model Logistic Regression
logreg = LogisticRegression(solver='liblinear', random_state=16)
ros = RandomOverSampler(random_state=16)

# Membuat pipeline
estimator = Pipeline([
    ('oversampling', ros),
    ('preprocess', transformer),
    ('model', logreg)
])
```

```python
# Menentukan ruang pencarian hyperparameter yang akan dievaluasi
hyperparam_space = [{
    'model__C': [0.01, 0.1, 1.0, 10,],       # Nilai regularisasi C (hyperparameter)
    'model__penalty': ['l1', 'l2'],         # Jenis penalti (l1 atau l2) untuk regularisasi
    'model__solver': ['liblinear'],         # Solver yang digunakan adalah 'liblinear' (untuk binary classification)
    #'model__max_iter': [none, 100, 200, 300],     # Batas maksimum iterasi untuk algoritma optimasi
    #'model__random_state': [none,16]
}]
```

```python
# Inisialisasi GridSearchCV untuk mencari kombinasi terbaik dari hyperparameter
grid = GridSearchCV(estimator, n_jobs=-1, param_grid=hyperparam_space, scoring='recall', cv=StratifiedKFold(n_splits=10, shuffle=True, random_state=16))
```

```python
# Melatih model menggunakan GridSearchCV untuk mencari hyperparameter terbaik
grid.fit(X_train,y_train)

# Menampilakn parameter terbaik dan nilai Recall
print("Best Recall score:", grid.best_score_)
print("Best hyperparameters:", grid.best_params_)
```

Hyperparameter terbaik untuk Logistic Regression dari hasil Gridsearch adalah `model__C` = 10, `model__penalty` = l1, `model__solver` = liblinear.

```python
# Mendapatkan model terbaik berdasarkan pencarian hyperparameter
best_model = grid.best_estimator_

# Melatih model terbaik dengan seluruh data pelatihan
best_model.fit(X_train, y_train)
```

*   Menambahkan fitur harga per layanan agar prediksi biaya yang dikeluarkan dan yang didapatkan bisa lebih jelas
*   Lakukan A/B testing atau backtesting pada model prediksi churn untuk mengevaluasi performanya di dunia nyata. Hasil dari pengujian ini dapat digunakan untuk memperbarui model secara periodik.
